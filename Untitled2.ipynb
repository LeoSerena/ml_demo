```python
import os

from tqdm import tqdm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Data Paramters

```python
M = 1000 # num users
N = 200 # num folders
D = 2 # vectors dimension
S = 100 # number of permissions per user
CATEGORIES = [ 'A', 'B', 'C', 'D' ]
PATH_WALKED = 'C:\\Users\\41763'
```

### Helpers Functions

```python
def normalize(X):
    return X / np.sqrt(np.sum(X * X, axis = 1)[:,np.newaxis])

def path_distance(p1, p2):
    l = len(os.path.commonprefix([p1, p2])) - 1
    return (p1[l:].count('\\') + p2[l:].count('\\')) ** 2
```

### Data Generation

```python
users = [ { 'name':i, 'cat' : np.random.choice(CATEGORIES) } for i in range(M) ]
df_users = pd.DataFrame(users)
user_names = df_users['name']
U = np.random.normal( 0, 1, (len(users), D) )
# U = normalize(U)
C = np.array(df_users['cat'])
C = np.tile(C, (len(C), 1))
users_distances = (C.T == C) * 1 

paths = [ os.path.join(r, fn) for r, _, f in os.walk(PATH_WALKED) for fn in f ]
np.random.shuffle(paths)
paths = paths[:N]
P = np.random.normal( 0, 1, (len(paths), D) )
# P = normalize(P)
K = np.array(paths)
K = np.tile(K, (len(K), 1))
paths_distances = np.array([ path_distance(p1, p2) for p1,p2 
    in tqdm(zip(
        K.reshape(len(P) ** 2),
        K.T.reshape(len(P) ** 2)
    ))
]).reshape(len(P), len(P))
paths_distances = paths_distances / np.max(paths_distances)
W = paths_distances / np.max(paths_distances)

permission_links = { u: np.random.choice(paths, size = S, replace = False) for u in df_users['name'] }
permissions_matrix = np.array([ (1 if x in permission_links[y] else 0) for x in paths for y in user_names ]).reshape(len(paths), len(user_names))
```

```python
plt.imshow(permissions_matrix)
plt.show()
plt.scatter(
    P[:,0],
    P[:,1]
)
plt.scatter(
    U[:,0],
    U[:,1]
)
plt.show()
```

### Losses

```python
def path_self_loss(k):
    mu = np.mean(np.tile(W[k], (D,1)).T * P, axis = 0)
    return np.sum((P[k] - mu) ** 2) 

def path_users_loss(k):
    mu = np.mean(U[np.argwhere(permissions_matrix[k] == 1).flatten()], axis = 0)
    return np.sum((P[k] - mu) ** 2)

def path_reg_loss(k):
    return P[k] @ P[k]

def path_loss(a, b, c, k):
    return a * path_self_loss(k) + b * path_users_loss(k) + c * path_reg_loss(k)

def paths_loss(*args):
    return np.mean([path_loss(*args, x) for x in range(N)])
```

```python
def user_self_loss(k):
    x = np.where(users_distances[k])
    y = np.where(2 * users_distances[k]-1)
    mu1 = np.mean(U[x], axis = 0)
    mu2 = np.mean(U[y], axis = 0)
    return np.sum((U[k] - mu1) ** 2) * len(x) / (len(x) + len(y)) - np.sum((U[k] - mu2) ** 2) * len(y) / (len(x) + len(y))

def user_paths_loss(k):
    mu = np.mean(P[np.argwhere(permissions_matrix.T[k] == 1).flatten()], axis = 0)
    return np.sum((U[k] - mu) ** 2)

def user_reg_loss(k):
    return U[k] @ U[k]

def user_loss(a, b, c, k):
    return a * user_self_loss(k) + b * user_paths_loss(k) + c * user_reg_loss(k)

def users_loss(*args):
    return np.mean([user_loss(*args, x) for x in range(M)])
```

```python
def loss(
    paths_weights = [1,1,1],
    users_weights = [1,1,1]
):
    return paths_loss(*paths_weights) + users_loss(*users_weights)
```

### Gradients

```python
def path_grad(a, b, c, k):
    p = P[k]
    mu = np.mean(np.tile(W[k], (D,1)).T * P, axis = 0)
    path_self_grad = a * (p - mu) * np.linalg.norm(p - mu)
    
    mu = np.mean(U[np.argwhere(permissions_matrix[k] == 1).flatten()], axis = 0)
    path_user_grad = b * (p - mu) * np.linalg.norm(p - mu)
    
    path_reg_grad = c * np.linalg.norm(p) * p
    
    return path_self_grad + path_user_grad + path_reg_grad

def user_grad(a, b, c, k):
    u = U[k]
    x = np.where(users_distances[k])
    y = np.where(2 * users_distances[k]-1)
    mu1 = np.mean(U[x], axis = 0)
    mu2 = np.mean(U[y], axis = 0)
    user_self_grad = a * (
        (u - mu1) * np.linalg.norm(u - mu1) - 
        (u - mu2) * np.linalg.norm(u - mu2)
    )
    
    mu = np.mean(P[np.argwhere(permissions_matrix.T[k] == 1).flatten()], axis = 0)
    user_user_grad = b * (u - mu ) * np.linalg.norm(u - mu)
    
    user_reg_grad = c * np.linalg.norm(u) * u
    
    return user_self_grad + user_user_grad + user_reg_grad
```

### Training

```python
# self, mix, reg
ap, bp, cp = [0.01, 1, 0]
au, bu, cu = [0.1, 1, 0.01]

losses = [np.inf, loss([ap,bp,cp],[au,bu,cu])]
```

```python
epochs = 200


lrp, lru = (0.2, 0.2)

P_idx = np.arange(len(P))
U_idx = np.arange(len(U))

# while losses[-1] < losses[-2] and len(losses) < 200:
for e in tqdm(range(epochs)):
    np.random.shuffle(P_idx)
    for p in P_idx:
        P[p] = P[p] - lrp * path_grad(ap, bp, cp, p)
    
    np.random.shuffle(U_idx)
    for u in U_idx:
        U[u] = U[u] - lru * user_grad(au, bu, cu, u)
    
    losses.append(loss([ap,bp,cp],[au,bu,cu]))
```
